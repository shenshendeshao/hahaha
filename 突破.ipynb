```python
class break_out():
    def __init__(self, df, window = 20, Pct = 0.20, box = 0.02):
        self.df = df
        self.win = window
        self.Pct = Pct
        self.box = box
        
    
    
    ##找到日K的箱体高点
    def find_highs(self):
        highs = []
        for i in range(len(df)):
            max_close_open =  max(df.loc[i, 'open'],df.loc[i, 'close'])
            highs.append({'date': df.loc[i, 'date'], 'max': max_close_open})
        return highs
    
    ###找出高点
    def check_tops(self):
        tops = []
        for i in range(2, len(df) - 2):
            if (self.find_highs()[i]['max'] >= self.find_highs()[i-1]['max'] and
                self.find_highs()[i]['max'] > self.find_highs()[i-2]['max'] and
                self.find_highs()[i]['max'] >= self.find_highs()[i+1]['max'] and
                self.find_highs()[i]['max'] > self.find_highs()[i+2]['max']):
                tops.append(self.find_highs()[i])
        tops.append(self.find_highs()[-1])
        tops.reverse()
        return tops
  
    

    
    ###以当日收盘价为基准向后检查是否突破
    def count_self(self):
        count = 0
        hits = []
        for i in self.check_tops():
            if (i['max'] <= self.check_tops()[0]['max'] and 
                i['max'] >= self.check_tops()[0]['max']*(1-self.box)):
                count += 1
                hits.append(i['date'])
                
            elif i['max'] > self.check_tops()[0]['max']:
                break
            
            else:
                pass
            
        if count >=3 :
            distance = len(df[df['date'] > hits[-1]][df['date'] < hits[0]])
        else:
            distance = 0
 
        return count,hits,distance

    ###以最近一个高点为基准向后检查是否突破
    def count_back(self):
        count = 0
        hits = []
        for i in self.check_tops():
            if (i['max'] <= self.check_tops()[1]['max'] and 
                i['max'] >= self.check_tops()[1]['max']*(1-self.box)):
                count += 1
                hits.append(i['date'])
                
            elif i['max'] > self.check_tops()[1]['max']:
                break
            
            else:
                pass
            
        if count >=3 :
            distance = len(df[df['date'] > hits[-1]][df['date'] < hits[0]] )
        else:
            distance = 0
 
        return count,hits,distance



    #起码有过三次尝试后突破
    def is_valid_1(self):
        if self.count_self()[0] >= 3 and self.count_self()[2] >= self.win:
            return True, self.count_self()[1]
        
        elif self.count_back()[0] >=3 and self.count_self()[2] >= self.win:
            return True, self.count_back()[1]
        
        else:
            return False, None
        
    

    #过去一个月涨幅不超过20%
        
    def is_valid_2(self):
        change = (df[-20:]['high'].max() - df[-20:]['low'].min())/df[-20:]['low'].min()
        
        if change >= self.Pct:
            return False
        
        else:
            return True

        
        
        
        
        
    #过去一个月内有过6%点以上阳线   
    def is_valid_3(self):                                    
        for i in range(1,20):
            if (df.iloc[-i]['close'] - df.iloc[-i]['open'])/df.iloc[-i]['open'] >= 0.06:    
                return True
            else:
                continue
            
        return False

    
    #年线上方
    #def is_valid_4(self):
    #    if len(df) == 250:
    #        if df.loc[249,'close'] >= df['close'].mean():
    #            return True
    #        else:
    #            return False
    #    else:
    #        return False
            
        
         
    
    def check(self):
        if self.is_valid_3() and self.is_valid_2():
            if self.is_valid_1()[0]:
                return True,self.is_valid_1()[1]
            else:
                return False,None
            
        else:
            return False,None
                               
```

```python
def get(stock_id):
    lg = bs.login()
    fields = "date,open,high,low,close"
    df_bs =bs.query_history_k_data(stock_id, fields, start_date='2022-03-01',frequency="D", adjustflag="2")
    
    date_list = []
    while (df_bs.error_code == '0') & df_bs.next():
        date_list.append(df_bs.get_row_data())

    result = pd.DataFrame(date_list, columns = df_bs.fields)
    result.close = result.close.astype('float64')               #收盘
    result.open = result.open.astype('float64')                 #开盘
    result.high = result.high.astype('float64')                 #最高
    result.low = result.low.astype('float64')                   #最低
    #result.volume = result.volume.astype('float64')             #成交量
    #result.pctChg = result.pctChg.astype('float64')             #涨跌百分比


    df = result[-80:]                                  #读取过去近半年数据（100天）
    df.reset_index(drop = True,inplace = True)
    ID = stock_id[3:]
    return df,ID
```

```python
SH = []
for ID in sh_id['code']:
    df = get(ID)[0]
    
    test = break_out(df)

    
    if test.check()[0] and len(SH)<5:
        
        SH.append([ID,test.check()[1]])
        
    elif len(SH)==5:
        break
```
